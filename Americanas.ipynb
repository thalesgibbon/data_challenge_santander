```python
#Função que procura um produto no mercado livre e retorna a lista de lojas na ordem que aparecem
def lista_vendedores_produto(nome_produto, lista_lojas, num_paginas):
    for i in range(num_paginas):
        url_inicial = 'https://www.magazineluiza.com.br/busca/' + nome_produto
        headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36", "Referer": url_inicial}
        html_content = requests.get(url_inicial, headers=headers).text
        soup = BeautifulSoup(html_content, "html")
        
        url_list = []
        for ul in soup.find_all("div", {"id": "productShowcaseSearch"}):
            for li in ul.find_all('li'):
                a = li.find('a')
                if a != None:
                   url_list.append(a['href'])

        for link in url_list:
            html_content = requests.get(link).text
            soup = BeautifulSoup(html_content, "html")
            a = soup.find("a", {"class": "seller-info-button js-seller-modal-button"})
            
            if a != None:
                lista_lojas.append(a['data-seller-razao-social'])

    return lista_lojas
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

nome_produto = "mp3"
lista_lojas = []
num_paginas = 1
pagina_pesquisa = ["", "/2", "/3", "/4", "/5"]

#escolher até 10 páginas
lista_lojas = lista_vendedores_produto(nome_produto, lista_lojas, num_paginas)

print(lista_lojas)
    
```
